```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from datetime import datetime, date,time,timedelta
from matplotlib.ticker import PercentFormatter
%matplotlib inline
```

```python
wd18 = pd.read_excel('wasaData.xlsx',sheet_name='2018')
wd17 = pd.read_excel('wasaData.xlsx',sheet_name='2017')
wd16 = pd.read_excel('wasaData.xlsx',sheet_name='2016')
wd15 = pd.read_excel('wasaData.xlsx',sheet_name='2015')
wd14 = pd.read_excel('wasaData.xlsx',sheet_name='2014')
wd13 = pd.read_excel('wasaData.xlsx',sheet_name='2013')
```

```python
def place (str):
    return int(str.split('/')[0])
def renameCat (str):
    return str[:3]+'-'+str[3:]
def renameCat2 (str):
    return str[:6]
def convert_timed(time):
    delta=timedelta(hours=time.hour,minutes=time.minute,seconds=time.second,microseconds=time.microsecond)
    return delta
def displayTime(x):
    y=timedelta(minutes=x)
    return str(y).split('.')[0]
```

```python
wds=[wd18,wd17,wd16,wd15,wd14,wd13]
```

```python
Tparticipants2=[]
for i in range(len(wds)):
    Tparticipants2.append(wds[i]['Name'].count())
Tparticipants2.reverse()
```

```python
wd18.rename(columns={20:"Category"}, inplace=True)
for i in range(1,len(wds)):
    wds[i].rename(columns={'Division':"Category"}, inplace=True)
```

```python
for i in range(len(wds)):
    wds[i]=wds[i].dropna(subset=['Place/Division','Category'])
```

```python

for wd in wds:
    wd['Place/Division']=wd['Place/Division'].apply(place)
    wd['Category']=wd['Category'].apply(renameCat)
```

```python
for i in range(len(wds)):
    wds[i] = wds[i][~wds[i]['Category'].isin(['C-1-20-154','M-1-55+','C-U-120','M-U-120','F-1-20-154','F-U-120','F-1-55+','M-1-20-154','CTM-10119','CTM-120154','CTM-155+','FTM-10119','FTM-120154','MTM-155+','MTM-10119','MTM-120154','M70-+OLY','C-1-55+'])]
```

```python
wds[5]['Category']=wds[5]['Category'].apply(renameCat2)
```

```python
#Dividing DataFrames in Male and Female
```

```python
wdM=[]
wdF=[]
for wd in wds:
    wdM.append(wd[~wd.Category.str.contains('F')])
    wdF.append(wd[~wd.Category.str.contains('M')])
```

```python
Fparticipants=[]
Mparticipants=[]
Tparticipants=[]
```

```python
for i in range(len(wds)):
    Fparticipants.append(wdF[i]['Name'].count())
    Mparticipants.append(wdM[i]['Name'].count())
    Tparticipants.append(wds[i]['Name'].count())
Fparticipants.reverse()
Mparticipants.reverse()
Tparticipants.reverse()
```

```python
col = ['Swim','T1','Bike','T2','Run','Time','Place/Division']
categ= ['M18-24','M25-29','M30-34','M35-39','M40-44','M45-49','M50-54','M55-59','M60-64','M65-69']
dfmall =[[] for _ in range(len(wdM))]
for j in range(len(dfmall)):
    for i in range(len(categ)):
        dfmall[j].append(wdM[j].loc[wdM[j]['Category']==categ[i],col])
```

```python

categF= ['F18-24','F25-29','F30-34','F35-39','F40-44','F45-49','F50-54','F55-59','F60-64']
dffall =[[] for _ in range(len(wdM))]
for j in range(len(dffall)):
    for i in range(len(categF)):
        dffall[j].append(wdF[j].loc[wdF[j]['Category']==categF[i],col])
```

```python
mav3 = [[] for _ in range(len(categ))]
fav3 = [[] for _ in range(len(categF))]
x=0
for j in range(len(categ)):
    for i in range(len(dfmall)):
        z =dfmall[i][j].loc[dfmall[i][j]['Place/Division']==3,['Time']]['Time']
        mav3[j].append(z)
for j in range(len(categF)):
    for i in range(len(dffall)):
        z=dffall[i][j].loc[dffall[i][j]['Place/Division']==3,['Time']]['Time']
        fav3[j].append(z)
        
```

```python
mv3 = [[] for _ in range(len(categ))]
fv3 = [[] for _ in range(len(categF))]
finalRM = []
finalRF = []
for j in range(len(mav3)):
    mav3[j] = pd.DataFrame(mav3[j])
    mav3[j].fillna(dt.time(0,0,0),inplace=True)
for j in range(len(fav3)):
    fav3[j] = pd.DataFrame(fav3[j])
    fav3[j].fillna(dt.time(0,0,0),inplace=True)
    
    
```

```python
for j in range(len(mav3)):
    for i in range(len(dfmall)):
        mv3[j].append(mav3[j].iloc[i,:].apply(convert_timed).max())
        for s in mv3[j]:
            if s == timedelta(0):
                mv3[j].remove(s)
    finalRM.append(pd.Series(mv3[j]).mean())
for j in range(len(fav3)):
    for i in range(len(dffall)):
        fv3[j].append(fav3[j].iloc[i,:].apply(convert_timed).max())
        for s in fv3[j]:
            if s == timedelta(0):
                fv3[j].remove(s)
    finalRF.append(pd.Series(fv3[j]).mean())
```

```python
finalRM=pd.Series(finalRM)
finalRF=pd.Series(finalRF)
```
